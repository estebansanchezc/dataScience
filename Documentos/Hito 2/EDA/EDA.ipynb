## Análisis Exploratorio Descriptivo

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use('seaborn')

import warnings
warnings.simplefilter('ignore')
```

```python
df = pd.read_csv('true_car_listings.csv', sep = ';')
df.head()
```

### Gráfico del precio de autos según sus marcas

```python
plt.figure(figsize = (25,10))
sns.boxplot(x='Make', y = 'Price', data = df)
plt.xticks(rotation = 90, fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Marca', fontsize = 18)
plt.ylabel('Precio', fontsize = 18)
plt.title('\nPrecio de automoviles por marca\n', fontsize = 25);
```
